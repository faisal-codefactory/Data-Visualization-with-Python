Seaborn is best with Pandas for Statistical Plots
Seaborn has three figure level plots
- relplot
   - scatterplot (kind='scatter')
   - lineplot (kind = 'line')
- displot
   - histplot
   - kdeplot
   - ecdfplot
   - rugplot
- catplot 
   - stripplot
   - swarmplot 
   - boxplot
   - violinplot
   - boxenplot
   - pointplot (kind = 'point')
   - barplot (kind = 'bar')
   - countplot (kind = 'count')
<p>
Different plots can be achieved by setting the 'kind' parameter.

Additionally, we can also used the axes level plots of seaborn.

```python
sns.displot(data=df, x= 'comb08', kind='hist')
```

```python
sns.displot(data=df, x= 'comb08', kind='hist', kde=True)
```

```python
sns.displot(data=df, x= 'comb08', kind='kde')
```

```python
# Empirical CDF (Cumulative density function)
# Tells the proporstion of data falling below each value.
# x-axis: values (sorted for this plot)
# y-axis: proportion of values we have seen fo far

sns.displot(data=df, x='comb08', kind='ecdf')
```

```python
sns.displot(data=df, x='comb08', kind='kde', rug=True)
```
